```python
from note_seq.protobuf import music_pb2

havana = music_pb2.NoteSequence()

# Add the notes to the sequence.
havana.notes.add(pitch=60, start_time=0.0, end_time=0.25, velocity=80)
havana.notes.add(pitch=64, start_time=0.25, end_time=0.50, velocity=80)
havana.notes.add(pitch=64, start_time=0.50, end_time=0.75, velocity=80)
havana.notes.add(pitch=60, start_time=0.75, end_time=1, velocity=80)
havana.notes.add(pitch=60, start_time=1, end_time=1.25, velocity=80)
havana.notes.add(pitch=69, start_time=1.25, end_time=1.50, velocity=80)
havana.notes.add(pitch=64, start_time=2, end_time=2.25, velocity=80)
havana.notes.add(pitch=62, start_time=2.25, end_time=2.5, velocity=80)
havana.notes.add(pitch=64, start_time=2.5, end_time=2.75, velocity=80)
havana.notes.add(pitch=65, start_time=2.75, end_time=3, velocity=80)
havana.notes.add(pitch=64, start_time=3, end_time=3.25, velocity=80) 
havana.notes.add(pitch=62, start_time=3.25, end_time=3.50, velocity=80) 
havana.notes.add(pitch=60, start_time=3.50, end_time=3.75, velocity=80) 
havana.notes.add(pitch=64, start_time=3.75, end_time=4, velocity=80) 
havana.notes.add(pitch=64, start_time=4, end_time=4.25, velocity=80) 
havana.notes.add(pitch=60, start_time=4.25, end_time=4.50, velocity=80) 
havana.notes.add(pitch=60, start_time=4.50, end_time=4.75, velocity=80) 
havana.notes.add(pitch=69, start_time=4.75, end_time=5, velocity=80)
havana.total_time = 5

havana.tempos.add(qpm=60);

# This is a colab utility method that visualizes a NoteSequence.
note_seq.plot_sequence(havana)

# This is a colab utility method that plays a NoteSequence.
note_seq.play_sequence(havana,synth=note_seq.fluidsynth)

# Here's another NoteSequence!
```

```python
print('Downloading model bundle. This will take less than a minute...')
note_seq.notebook_utils.download_bundle('basic_rnn.mag', '/content/')

# Import dependencies.
from magenta.models.melody_rnn import melody_rnn_sequence_generator
from magenta.models.shared import sequence_generator_bundle
from note_seq.protobuf import generator_pb2
from note_seq.protobuf import music_pb2

# Initialize the model.
print("Initializing Melody RNN...")
bundle = sequence_generator_bundle.read_bundle_file('/content/basic_rnn.mag')
generator_map = melody_rnn_sequence_generator.get_generator_map()
melody_rnn = generator_map['basic_rnn'](checkpoint=None, bundle=bundle)
melody_rnn.initialize()

print('ðŸŽ‰ Done!')
```

```python
# Model options. Change these to get different generated sequences! 

input_sequence = three_little_birds # the sequence that will be used
num_steps = 128 # num_steps is how long each melody will be in 16th steps (128 steps = 8 bars)
temperature = 1.5 # the higher the temperature the more random the sequence.

# Set the start time to begin on the next step after the last note ends.
last_end_time = (max(n.end_time for n in input_sequence.notes)
                  if input_sequence.notes else 0)
qpm = input_sequence.tempos[0].qpm #the tempo that is set before
seconds_per_step = 60.0 / qpm / melody_rnn.steps_per_quarter 
total_seconds = num_steps * seconds_per_step #equals end_time

generator_options = generator_pb2.GeneratorOptions()
generator_options.args['temperature'].float_value = temperature
generate_section = generator_options.generate_sections.add(
  start_time=last_end_time + seconds_per_step,
  end_time=total_seconds)

# Ask the model to continue the sequence.
sequence1 = melody_rnn.generate(input_sequence, generator_options)

note_seq.plot_sequence(sequence1)
note_seq.play_sequence(sequence1, synth=note_seq.fluidsynth)
```

```python
note_seq.sequence_proto_to_midi_file(sequence1, 'three_little_birds.mid')

files.download('three_little_birds.mid')
```
